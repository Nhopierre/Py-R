## Comment créer un fichier son au format wave avec python

### Présentation d'un fichier son

Un fichier au format wave est un fichier qui contient les informations nécessaires pour stocker un signal d'origine sonore.<br>
Il est caratérisé par sa fréquence d'échantillonnage et sa quantification.

Exemple de fichier son édité avec Audacity. L'axe est gradué en minutes:secondes.
L'extraite dure à peine quelques millisecondes
